### Data Frame Arithematic
Since DataFrames are collections of series we can perform series like operations on them, an important feature to keep in mind is the data alignment feature of pandas. lets explore Data frame arithematics under the following:
+ Within the DataFrame
+ DataFrame vs Scalar
+ DataFrame vs DataFrame

#### Within the DataFrame
We can perform simple arithematics between columns of the DataFrame, this is done by selecting the desired columns arnd using the regular Python arithematic operators.

```python
test1 = pd.Series([12,15,14,11,12,18])
test2 = pd.Series([16,12,11,13,12,13])
exam = pd.Series([55,39,38,55,54,45])
student =  ["dan","ib","rak","tam","nik","salv"]

ser_dict = {"student":student,"test1":test1,"test2":test2,"exam":exam}
data = pd.DataFrame(ser_dict)

data
```

```python
# ADDING COLUMNS WITHIN THE DATA FRAME 
data["Total"] = data["test1"] + data["test2"] + data["exam"]
```

```python
data
```

#### DataFrame vs Scalar

```python
# DIVIDING A COLUMN BY A SCALAR
data["Points"] = data["Total"] / 100
```

```python
data
```

### eval method for efficient Operations. (Optional)
The eval() function in pandas uses strings expressions to efficiently compute operations using DataFrames. It supports:
+ Arithematic Operators
+ Compairson Operators 
+ Bitwise operators  (& and | bitwise operators)
+ Object Attributes and indices (access to object attributes via **object.attr** syntax and indexes via **obj[index]** syntax

> **syntax: pd.eval(expr,target,inplace)**

Parameter | Description
:- | :-
expr | str The expression to evaluate. This string cannot contain any Python statements
target | object, optional, default None. This is the target object for assignment. It is used when there is variable assignment in the expression. If so, then `target` must support item assignment with string keys, and if a copy is being returned, it must also support `.copy()`.
inplace | bool, default False If `target` is provided, and the expression mutates `target`, whether to modify `target` inplace. Otherwise, return a copy of `target` with the mutation.

**WARNING** *you should not use eval() for simple expressions or for expressions involving small DataFrames. In fact, eval() is in many orders of magnitude slower for smaller expressions/objects than plain olâ€™ Python. A good rule of thumb is to only use eval() when you have a DataFrame with more than 10,000 rows.*

```python
# eval with string arithematic expresssion
pd.eval("(df1 * df2) / 100")
```

```python
df1
```

```python
# eval with string arithematic expresssion
pd.eval("E = (df1.A * 50) / 100", target=df1, inplace=False)
```

```python
# eval with string arithematic expresssion
pd.eval("df1 * df2")
```

```python
df1
```

```python
# eval with string compairson expresssion
pd.eval("df1 > 5")
```

```python
# eval with string compairson and bitwise operators 
pd.eval("(df1> 5) | (df2 <4)")
```

```python
# eval with string airthematic expresssion
pd.eval("df1.A + df1.B")
```

**NB** we can also call the eval method off a DataFrame object, doing it this way we can just call the column names as variables, and when you want to reference a local variable with a unique name space you use an @ before the variable name `@variable_name`

```python
# Assigning a column with target
df1.eval("F = A + B * C",target=df1)
```

```python
num = 100
```

### Operating on null Values
As we have encountered missing data in some previous examples, we can now say that missng data is common in most data science analysis applications. Pandas provides us with the ability to deal with such missing data easily. 
Pandas uses the floating point value **NaN** (Not a Number) to represent missing data in both floating as wellas in non floating point arrays. It serves as an indicator of the presence of missing data.so it can easily be detected.
+ Dectecting null values
+ Dropping null values 
+ filling null values 

**Dectecting null values**  
Pandas Data structure has three usefull methods for detecting null data: 

Method | Description
:- | :-
**isna()** or **isnull()**  | Detect missing values.Return a boolean same-sized object indicating if the values are NA. NA values, such as None or :attr:`numpy.NaN`, gets mapped to Truevalues.
**notna** or **notnull()**  | Detect existing (non-missing) values. Return a boolean same-sized object indicating if the values are not NA. Non-missing values get mapped to True.

either one will return a Boolean mask over the data.

```python
product = df1 * df2
```

**Dropping null values**  
*In certain situations, you might want to delete rows or even columns that have missing values. You can do this with **.dropna()** In this case, .dropna() simply deletes the row with nan, including its label. The result of this does not affect the DataFrame directly, we have the option to do it in place (optional parameter) which actually drops the rows from the data frame inplace:*

> **syntax: df.dropna(axis=0, how='any', thresh=None, inplace=False)**

Parameter | Description
:- | :-
axis | {0 or 'index', 1 or 'columns'}, default 0. Determine if rows or columns which contain missing values are removed.
| 0 or 'index' : Drop rows which contain missing values.
| 1 or 'columns' : Drop columns which contain missing value.
how | {'any', 'all'}, default 'any'.Determine if row or column is removed from DataFrame, when we have at least one NA or all NA.
| 'any' : If any NA values are present, drop that row or column.
| 'all' : If all values are NA, drop that row or column.
thresh | int, optional. Require that many non-NA values.
inplace |bool, default False. If True, do operation inplace and return None.

**NB** 
+ The thresh parameter lets you specify a **minimum** number of **non-null** values for the rows/columns to be kept. (thresh=N requires that a column has at least N non-NaNs to survive,Any row/column with the number of **non-NA values < thresh** value is removed irrespective of other parameters passed).  
+ While the **how** parameter  if set to **"any"** will drop any row or column containing a null value, but if set to **"all"** will drop only rows or columns that are filled with null values.

```python
product["G"] = np.nan
product.loc["y",:] = np.nan
product
```

```python
# Column G was dropped because all the elements in that column were NaN
product.dropna(axis="columns", how= "all")
```

```python
# All columns have been dropped because the had NaNs
product.dropna(axis="columns", how= "any")
```

```python
# All rows have been dropped becaused each contained a NaN (how="any")
product.dropna(axis="index", how="any")
```

```python
# row Y was dropped because all its elements were NaNs
product.dropna(axis="index", how="all")
```

```python
# All columns with Non-NAN less than thress were dropped 
product.dropna(axis="columns",thresh=5)
```

```python
# Rows with Non-NAN less than thresh was dropped
product.dropna(axis="index", thresh=5)
```

```python
# All rows with Non-NAN less than 6 was dropped
product.dropna(axis="index", thresh=6)
```

**NB** for finer contol over dropping NANs use a combination of axis and thresh parameters, and dont forget to set inplace to True if you need the changes to reflect on the DataFrame.

**filling null values**  
There are times when we will rather replace these null values than drop them. The replace value may be a single number, or an interpolation from the good values or may be based on some plilosophy. One of the most convenient methods is .fillna(). You can use it to replace missing values with:
+ Specified values
+ The values above the missing value
+ The values below the missing value

syntax: ``df.fillna(value=None,method=None,axis=None,inplace=False,limit=None,downcast=None,)``

Parameter | Description
:- |:-
value | scalar, dict, Series, or DataFrame. Value to use to fill missing values (e.g. 0), alternately a dict/Series/DataFrame of values specifying which value to use for each index (for a Series) or column (for a DataFrame).  Values not in the dict/Series/DataFrame will not be filled. This value cannotbe a list.
method | {'backfill', 'bfill', 'pad', 'ffill', None}, default None. Method to use for filling holes in reindexed Series pad / ffill: propagate last valid observation forward to next valid backfill / bfill: use next valid observation to fill gap.
axis | {0 or 'index', 1 or 'columns'} Axis along which to fill missing values.
inplace | bool, default False If True, fill in-place. Note: this will modify any other views on this object (e.g., a no-copy slice for a column in a DataFrame).
limit | int, default None If method is specified, this is the maximum number of consecutive NaN values to forward/backward fill. In other words, if there is a gap with more than this number of consecutive NaNs, it will only be partially filled. If method is not specified, this is the maximum number of entries along the entire axis where NaNs will be filled. Must be greater than 0 if not None.
downcast | dict, default is None. A dict of item->dtype of what to downcast if possible, or the string 'infer' which will try to downcast to an appropriate equal type (e.g. float64 to int64 if possible).

```python
product
```

```python
# filling with a scalar value
product.fillna(value =10000)
```

```python
# filling with the mean value 
product.fillna(value =product["A"].mean())
```

```python
# Foward fill- (propagate the previous value foward)
product.fillna(method="ffill",axis=1)
```

```python
product
```

```python
# Backward fill- (propagate the next value backward)
product.fillna(method="bfill",axis=0)
```

**NOTE** When the next or previous value is not available, the NA value remains  
Another popular option is to apply interpolation and replace missing values with interpolated values. You can do this with .interpolate():




```python
# Using interpolation to fill NAN
product.interpolate()
```

mean() | Mean of non-NA values
